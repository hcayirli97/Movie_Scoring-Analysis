```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
movies = pd.read_parquet("files/movies.parquet")
ratings = pd.read_parquet("files/ratings.parquet")
```

```python
movies.head(1)
```

```python
ratings.head(1)
```

```python
data = pd.concat([movies, ratings], axis=1)
data.head(1)
```

```python
data.info()
```

```python
fig = data.groupby('year').mean()['avg_vote'].plot(title = "Avg Vote", figsize= (15,8))
fig.get_figure().savefig("results/year_avgvote.png")
```

```python
fig = data.groupby('year').mean()['total_votes'].plot(title = "Total Votes" , figsize= (15,8))
fig.get_figure().savefig("results/year_totalvotes.png")
```

```python
data['Score'] = data['avg_vote'] * data['total_votes']
best_data = data.sort_values(by='Score', ascending=False)
best_data.head()
```

```python
fig = best_data[['original_title','Score']].iloc[:20].sort_values(by='Score', ascending=True).plot.barh(x = "original_title",figsize= (10,8),ylabel = "Score",color = "g")
fig.get_figure().savefig("results/best_score.png")
```

```python
fig = data.groupby('year').mean()['Score'].plot(title = "Score", figsize= (15,8))
fig.get_figure().savefig("results/year_score.png")
```

```python
genre_data = data.copy()
genre_data['genre'] = genre_data['genre'].str.split(',')
genre_data = genre_data.explode('genre')
genre_data['genre'] = genre_data['genre'].str.strip()
fig = genre_data.groupby('genre')['avg_vote'].mean().sort_values(ascending = False).plot.bar(figsize= (20,8),ylabel = "Avg Vote")
fig.get_figure().savefig("results/genre_avgvote.png")
```

```python
temp = genre_data.groupby('genre')['genre'].count().to_frame(name = "counts")
temp = temp.sort_values(by='counts', ascending=False)
fig = temp[temp['counts'] > 1500].plot.pie(y="counts", legend = False, figsize= (10,10),autopct='%.2f')
fig.get_figure().savefig("results/counts_genre.png")
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
```

```python
genres = list(genre_data.genre.unique())
```

```python
trends = {}
for genre in genres:
    train_df =  genre_data[genre_data['genre'] == genre].sort_values("year",ascending = False)[["Score","year"]].groupby('year').mean()
    X = pd.DataFrame(train_df.index)
    y = train_df['Score']
    lin = LinearRegression()
    lin.fit(X,y)
    trends[genre] = lin.predict(X)
```

```python
from labellines import labelLine, labelLines
import matplotlib.pyplot as plt
```

```python
fig, axs = plt.subplots(4, 5)
fig.set_figheight(25)
fig.set_figwidth(25)
fig.suptitle('Trendline - Score Graph', fontsize=35)
k = 0
for i in range(4):
    for j in range(5):
        key_name = genres[k]
        axs[i][j].plot(genre_data[genre_data['genre'] == key_name].sort_values("year",ascending = False).groupby('year').mean()['Score'])
        axs[i][j].plot(genre_data[genre_data['genre'] == key_name].sort_values("year",ascending = False)[["Score","year"]].groupby('year').mean().index, np.array(trends[key_name]))
        axs[i][j].set_title(key_name)
        k+= 1       
plt.savefig('results/Trendline-Score Graph.png', bbox_inches='tight')
```
